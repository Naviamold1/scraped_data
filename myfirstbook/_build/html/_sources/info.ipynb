## კითხვები რომლებიც ნაღდად მოდიან

1. **ბუნებაში ყველა მაკროსკოპული პროცესი:**

   - შექცევადია
   - [x] (ბ) შეუქცევადია
   - შექცევადიცაა და შეუქცევადიც გააჩნია პირობებს.

2. **რაში მდგომარეობს თერმოდინამიკის პირველი და მეორე კანონები?**

   ერთი მდგომარეობიდან მეორეში გადასვლისას სისტემის ∆U შინაგანი ენერგიის ცვლილება ტოლია გარე ძალთა A მუშაობისა და სისტემაზე გადაცემული Q სითბოს რაოდენობის ჯამისა:
   ∆U = A+Q
   ენტროპია იზოლირებულ სისტემაში არ იკლებს. სითბო არ შეიძლება თვითნებურად გადავიდეს ცივი სხეულიდან ცხელზე სამუშაოს დახმარების გარეშე.

3. **რა პირობებია საჭირო,რომ იმუშაოს სითბურმა ძრავამ?**
   სითბურ მანქანას სჭირდება: სახურებელი (საწვავი), მუშათსხეული, მაცივარი.

4. **სითბური მანქანა რომლის მ.ქ. კოეფიციენტი 60%-ია, გამახურებლიდან ერთი ციკლის განმავლობაში იღებს 100ჯ. სითბოს და ასრულებს სასარგებლო მუშაობას არის:**

   - () 40ჯ
   - [x] (ბ) 60ჯ
   - () 100ჯ
   - () 160ჯ

5. **მაქსიმალური შეღწევისუნარიანობა გააჩნია:**

   - () α ნაწილაკებს
   - () β ნაწილაკებს
   - [x] (გ) γ ნაწილაკებს

6. **ორთქლის მანქანასა და ტურბინაში ორთქლის ტემპერატურა მუშაობის შესრულების პროცესში:**

   - [x] (ბ) მცირდება

7. **ქიმიური ელემენტის ატომგულის თვითნებურ დაშლას, რასაც თან სდევს სხვადასხვა ნაწილაკების გამოსხივება, ეწოდება:**

8. **რაიმე მასის მქონე ნივთიერებისათვის საჭირო სითბოს რაოდენობა განისაზღვრება:**

## კითხვები რომლებიც დიდი ალბათობით მოვლენ

1. **რა თერმოდინამიკური სიდიდებით ხასიათდება თერმოდინამიკური სისტემა?**

   - ტემპერატურა, მოცულობა, წნევა

2. **რომელია კონდენსაციის პროცესი?**

   - აირიდან თხევადში (Gas to Liquid)

3. **შინაგანი ენერგია რითი არ შეიძლება შეიცვალოს?**

   - ეკვატორიდან პოლუსზე გადატანით

4. **ნაწილაკების ქაოსური მოძრაობას რა ეწოდება?**

   - სითბური მოძრობა

5. **აბსოლუტური ნული არის?**

   - -273°C

6. **გადაიყვანეთ X°C კელვინებში**

   - პასუხი უდრის: X−273

7. **თუ წნევა მუდმივია როგორია მოცულობის დამოკიდებულება ტემპერატურაზე?**

   $$V = V~0~d$$
   $$d=\frac{1}{273}$$

   - გრაფიკი (ახლა დაცინვა არ დამიწყოთ):
     
     ![image](https://img001.prntscr.com/file/img001/cpUfMV7zTnOJdDwgrPeIrQ.png)

8. **შინაგანი ენერგიის ცვლილება გადაცემულ სითბო და სამუშაო რამდენია შინაგანი ენერგიის ცვლილება?**

   - ეს არ ვიცი (თუ იცით მომწერეთ)

9. **რა მასის ნავთი შეიძლება გავათბოთ 1°C ნავთის კუთრს სითბოტევადობა** (ვერ გავიგე აქ რა გიწერია, მარი)

   - სცადეთ ეს ფორმულება idk
     $$Q = qm$$
     $$Q = cm(∆t)$$
